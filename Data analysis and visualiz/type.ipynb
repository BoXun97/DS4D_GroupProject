```python
import pandas as pd
datalist = pd.DataFrame(
    {
        'labels': ['Intra ', 'IntraLocal', 'InterIntra','Inter'],
        'Africa': [105, 15, 12, 4],
        'Europe and Eurasia': [50, 2, 34,0],
        'Asia and Pacific': [45, 0, 3, 0],
        'Middle East and North Africa': [11, 2, 7, 1],
        'Americas': [18, 0, 0,0],
        'Cross-regional': [0, 0, 2, 0],
    },
).set_index('labels')

fig = plt.figure(
    FigureClass=Waffle,
    plots={
        '611': {
            'values': datalist['Africa'] ,
            'labels': ["{0} ({1})".format(n, v) for n, v in datalist['Africa'].items()],
            'legend': {'loc': 'upper left', 'bbox_to_anchor': (0.5, 1), 'fontsize': 8},
            'title': {'label': 'Number of conflicts degrees--Africa', 'loc': 'left'}
        },
        '612': {
            'values': datalist['Europe and Eurasia'],
            'labels': ["{0} ({1})".format(n, v) for n, v in datalist['Europe and Eurasia'].items()],
            'legend': {'loc': 'upper left', 'bbox_to_anchor': (1.2, 1), 'fontsize': 8},
            'title': {'label': 'Number of conflicts degrees--Europe and Eurasia', 'loc': 'left'}
        },
        '613': {
            'values': datalist['Asia and Pacific'],
            'labels': ["{0} ({1})".format(n, v) for n, v in datalist['Asia and Pacific'].items()],
            'legend': {'loc': 'upper left', 'bbox_to_anchor': (1.3, 1), 'fontsize': 8},
            'title': {'label': 'Number of conflicts degrees--Asia and Pacific', 'loc': 'left'}
        },
        '614': {
            'values': datalist['Middle East and North Africa'],
            'labels': ["{0} ({1})".format(n, v) for n, v in datalist['Middle East and North Africa'].items()],
            'legend': {'loc': 'upper left', 'bbox_to_anchor': (1.4, 1), 'fontsize': 8},
            'title': {'label': 'Number of conflicts degrees--Middle East and North Africa', 'loc': 'left'}
        },
         '615': {
            'values': datalist['Americas'],
            'labels': ["{0} ({1})".format(n, v) for n, v in datalist['Americas'].items()],
            'legend': {'loc': 'upper left', 'bbox_to_anchor': (1.5, 1), 'fontsize': 8},
            'title': {'label': 'Number of conflicts degrees--Americas', 'loc': 'left'}
        },
         '616': {
            'values': datalist['Cross-regional'],
            'labels': ["{0} ({1})".format(n, v) for n, v in datalist['Cross-regional'].items()],
            'legend': {'loc': 'upper left', 'bbox_to_anchor': (1.5, 1), 'fontsize': 8},
            'title': {'label': 'Number of conflicts degrees--Cross-regional', 'loc': 'left'}
        },
    },
    rows=5,
    colors=("#2196f3", "#ff5252", "#999999","#497998"),  # Default argument values for subplots
    figsize=(20, 20)  # figsize is a parameter of plt.figure
)
```
