```python
with open('con-664.csv','r') as file:
    f = file.readlines()

text = []
flag = False
tag = False
for i in range(len(f)):
    for j in range(len(f)):
        if f[i] == f[j] and i !=j:
            for k in range(len(text)):
                if f[j]  == text[k]:
                    flag = True
                    tag = True
            if flag:
                flag = False
                continue
            text.append(f[j])
            print('{}and{}'.format(i,j)) #Output the position of repeated word lines
            print(f[j])  #Output repeated words
            
    if tag:
        tag = False
        continue
    #After deleting the repeated words, write a new document
    with open('word.txt','a') as ft:
        ft.writelines(f[i]) 
```
